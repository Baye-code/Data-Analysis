#### Best Month's Sales ?

```python
Sales_Year_2019.columns
```

```python
Sales_Year_2019['Price Each'] = pd.to_numeric(Sales_Year_2019['Price Each'])
Sales_Year_2019['Quantity Ordered'] = pd.to_numeric(Sales_Year_2019['Quantity Ordered'])
```

```python
Sales_Year_2019['Total Sales by Month'] = Sales_Year_2019['Price Each'] * Sales_Year_2019['Quantity Ordered']
```

```python
Sales_Year_2019.rename(columns={'Total Sales by Date':'Total Sales'}, inplace=True)
Sales_Year_2019.head()
```

```python
Monthly_Sales = Sales_Year_2019.groupby('Month').sum()
Monthly_Sales
```

```python
months = range(1, 13)

plt.style.use('fivethirtyeight')

plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.tight_layout()
plt.xticks(months)

plt.bar(months, Monthly_Sales['Total Sales by Month'])
plt.plot(months, Monthly_Sales['Total Sales by Month'], linestyle='--', color='red')
plt.show()
```

#### Highest Number of Sales per City

#### Maximizing Likelihood of buying product from customers ** Time Display **

```python
Sales_Year_2019['Order Date'] = pd.to_datetime(Sales_Year_2019['Order Date'])
```

```python
Sales_Year_2019['Hour'] = Sales_Year_2019['Order Date'].dt.hour
Sales_Year_2019['Minute'] = Sales_Year_2019['Order Date'].dt.minute
Sales_Year_2019.head()
```

```python
hours = [hour for hour, df in Sales_Year_2019.groupby('Hour')]

plt.plot(hours, Sales_Year_2019.groupby(['Hour']).count())

plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()

plt.show();
```

#### What products are most often sold together

```python
df = Sales_Year_2019[Sales_Year_2019['Order ID'].duplicated(keep=False)]
```

```python
df.head()
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head(5)
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(10)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```
