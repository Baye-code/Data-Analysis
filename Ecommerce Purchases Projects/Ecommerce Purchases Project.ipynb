```python
# Method 2
```

```python
def fun():
    count = 0
    for date in data['CC Exp Date']:
        if date.split('/')[1] == '20':
            count = count + 1
    print(count)
```

```python
fun()
```

## Top 5 most email providers

```python
# Method 1 
```

```python
def email_providers():
    famous_email_providers = []
    for email in data['Email']:
        famous_email_providers.append(email.split('@')[1])
    
    famous_email_providers_Series = pd.Series(famous_email_providers)
    print(famous_email_providers_Series.value_counts().head())
```

```python
email_providers()
```
